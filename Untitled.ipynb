```python
import serial
import time

BASE_STATION_PORT = "COM9" 

print(f"Opening simple line listener on {BASE_STATION_PORT}...")
with serial.Serial(BASE_STATION_PORT, 115200, timeout=1) as ser:
    ser.reset_input_buffer()
    print("Listening... Wiggle your remote sensor or press its reset button now.\n")
    
    while True:
        if ser.in_waiting:
            # Grab the raw line exactly as the hardware dumps it
            raw_line = ser.readline().decode('utf-8', errors='ignore').strip()
            
            if raw_line:
                # Just print it raw! No splitting, no dictionaries, no crashes.
                print(f"HW DUMP: {raw_line}")
                
        time.sleep(0.01)
```
